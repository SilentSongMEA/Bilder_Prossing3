```python
drowing=False
start_point=(0,0)
def drow_rectangle(event,x,y,flags,param):
    global start_point,drowing
    if event==cv2.EVENT_LBUTTONDOWN:
        drowing=True
        start_point=x,y
    elif event==cv2.EVENT_MOUSEMOVE:
        if drowing==True:
            cv2.rectangle(img,start_point,(x,y),(255,255,255),-1)
    elif event==cv2.EVENT_LBUTTONUP:
        drowing=False
img=np.zeros((512,512,3),np.uint8)
cv2.imshow("image",img)
cv2.setMouseCallback("image",drow_rectangle)
while True:
    cv2.imshow("image",img)
    if cv2.waitKey(1)&0xFF==27:
        break
cv2.destroyAllWindows()
plt.imshow(img[:,:,::-1])
```

```python
img=np.zeros((512,512,3))
drowing=False
ix,iy=0,0
def drow(event,x,y,flags,param):
    global ix,iy,drowing
    if event==cv2.EVENT_LBUTTONDOWN:
        drowing=True
        ix=x
        iy=y
        
    elif event==cv2.EVENT_MOUSEMOVE:
        if drowing==True:
            cv2.line(img,pt1=(ix,iy),pt2=(x,y),color=(255,255,255),thickness=3)
            ix=x
            iy=y
    elif event==cv2.EVENT_LBUTTONUP:
        drowing=False
img=np.zeros((512,512,3),np.uint8)
cv2.imshow("image",img)
cv2.setMouseCallback("image",drow)
while True:
    cv2.imshow("image",img)
    if cv2.waitKey(1)&0xFF==27:
        break
cv2.destroyAllWindows()
plt.imshow(img[:,:,::-1])
```
